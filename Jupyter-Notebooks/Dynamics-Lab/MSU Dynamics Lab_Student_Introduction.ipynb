## We can do <i>$nested$</i> for-loops. 

### The outer loop <i>i</i> will start with its first entry, 0 and then the inner loop <i>j</i> will run through all of it's entries before the <i>i</i> moves onto its next entry, in which <i>j</i> will run through all of its entries again, etc:

```python
for i in range(0,4):
    for j in range(0,6):
        print("i:",i)
        print("j:",j)
        print("i+j =",i+j,"\n")
    print("Done with all j's; moving onto next i-value:\n")
print("Done.")
```

# Functions!

### We can make functions that we can call over and over again. The arguments needed for the function are in the parenthesis

There are a couple main parts to a function in python:
* $def$: Define a function
* Arguments: variables required for the function (think $x$ in $f(x) = x^{2}$)
* Indented lines for actions: anything that you want to run in the function must be indented below the $def$
* $return$: Generate something when the function is called


### Make a function that will add 4 to any number we feed into the function

```python
def Add_to_4(number):
    new_number = number+4.
    print("4.0 +",number,"=",new_number)
    return new_number
```

```python
Add_to_4(5.)
```

```python
# We can assign the resultant value of the function to a variable as well.
New_Num = Add_to_4(5.)
print(New_Num)
print("New_Num type:",type(New_Num))
print(New_Num + 30.)
```

## We can even run a function over and over in our newly aquired for-loops

```python
for i in range(5): # the range() argument must be integers!!
    print("index number: ",i)
    
    # Remember i is an int, we're looking for a float for the function
    print("Function output: ",Add_to_4(float(i)),"\n")
    
```

# Arrays!
## Time to work with some very flexible and widely used objects called arrays

```python
# xarray will allow us so many cool and useful features when working with multidimensional data arrays
import xarray as xr
```

### We can define a random number to set the size of our dimension in our 1d array:

```python
test_num = 14
```

### Let's make a 1d array with number of entries equal to 14:

```python
# Initate an empty 1d array to populate with Coriolis factors for each lat

test = xr.DataArray(np.zeros((test_num))) # test will be a 1d array with test_num number of entries

for i in range(test_num):  # same thing as range(0,14)
    test[i]=i # Populating each ith index with the value for i, with i being an integer
        
print("done.")
print("Shape of array:",test.shape)
print("Number of dimensions in array:",len(test.shape))
print(test.values,"\n")
print("So, this is a",str(len(test.shape))+"-d array with length of",test.shape[0],"for its only dinemsion")
```

### Let's make a 2d array with number of entries equal to 14 in each dimension:

```python
test2 = xr.DataArray(np.zeros((test_num,test_num))) # Same as before but now we have 2 dims with 14 entries each

for i in range(test_num):
    for j in range(test_num):
        test2[i,j]=i
        
print("done.")
print("Shape of array:",test2.shape)
print("Number of dimensions in array:",len(test2.shape))
```

### Let's make a 4d array with number of entries equal to 14 in each dimension:

```python
test4 = xr.DataArray(np.zeros((test_num,test_num,test_num,test_num)))

for i in range(test_num):
    for j in range(test_num):
        for k in range(test_num):
            for l in range(test_num):
                test4[i,j,k,l]=i
        
print("done.")
```

```python
print("Shape of array:",test4.shape)
print("Number of dimensions in array:",len(test4.shape))
print("Shape of test4[0:2,:,:,:]:",(test4[0:2,:,:,:].shape))
```
